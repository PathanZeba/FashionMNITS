```python
import json
import matplotlib.pyplot as plt
import pandas as pd
```

```python
with open("results/metrics_dense.json") as f:
    dense_metrics = json.load(f)

with open("results/metrics_cnn.json") as f:
    cnn_metrics = json.load(f)

df = pd.DataFrame([dense_metrics, cnn_metrics], index=["Dense NN", "CNN"])
df
```

```python
plt.figure(figsize=(6,4))
plt.bar(df.index, df["acc"], color=["skyblue","orange"])
plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy")
plt.ylim(0,1)
for i, v in enumerate(df["acc"]):
    plt.text(i, v+0.01, f"{v:.3f}", ha="center", fontsize=10)
plt.show()
```

```python
print("Comparative Analysis Summary")
print(f"Dense NN Accuracy: {dense_metrics['acc']:.4f}")
print(f"CNN Accuracy:      {cnn_metrics['acc']:.4f}")

if cnn_metrics["acc"] > dense_metrics["acc"]:
    print("CNN outperforms Dense NN on Fashion-MNIST.")
else:
    print("Dense NN performed better (unexpected, check training setup).")
```
